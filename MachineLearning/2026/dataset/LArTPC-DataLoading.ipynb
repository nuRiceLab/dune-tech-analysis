# LArTPC Particle Images

This notebook mirrors the data-loading part of the MNIST tutorial, but uses the tiny LArTPC particle-image dataset.

```python
# Import drawing tools
import matplotlib.pyplot as plt
%matplotlib inline

# Import path utilities and torch data-loading utilities
from pathlib import Path
import sys
import torch
from torch.utils.data import DataLoader

# Import the local MNIST-like LArTPC dataset
from lartpc_particles import LArTPCParticles
```

## 1. LArTPC Dataset

The dataset follows the same basic access pattern as `torchvision.datasets.MNIST`: create a dataset object, check its length, and access entries by index. Each entry is a tuple containing an image tensor and an integer label.

```python
# Specify where you want the dataset to land
LOCAL_DATA_DIR = 'mlar-dataset'

# Fill this in once the dataset file is hosted.
# You can also use a local path, e.g. DATA_URL='/path/to/lartpc_particles.pt'
DATA_URL = 'https://s3df.slac.stanford.edu/data/neutrino/dune-tech-analysis/mlar/raw/lartpc_particles.pt'

# Create the dataset. If DATA_URL is set, download/process into LOCAL_DATA_DIR.
dataset = LArTPCParticles(
    LOCAL_DATA_DIR,
    train=True,
    download=DATA_URL is not None,
    url=DATA_URL,
)

print('Using data cache:', LOCAL_DATA_DIR)
```

```python
# Length of the dataset
print(len(dataset))

# Type of the first element in the dataset
print(type(dataset[0]))
```

Each dataset entry is a `(data, label)` tuple. The data tensor has shape `[1, 28, 28]`, matching the MNIST convention: one grayscale channel and a 28-by-28 image.

```python
# Access a specific entry in the dataset
ENTRY = 0
data, label = dataset[ENTRY]

# Show the type/shape of data and the label associated with it
print('Type of data  :', type(data), 'shape', data.shape)
print('Type of label :', type(label), 'value', label, dataset.idx_to_class[label])

# Use matplotlib to draw the data as an image
image = data.view(data.shape[1:])
plt.imshow(image, cmap='gray', origin='lower')
plt.title(f'{label}: {dataset.idx_to_class[label]}')
plt.axis('off');
```

## 2. Streaming Data Using `DataLoader`

As with MNIST, the dataset can be wrapped in a `DataLoader` to produce mini-batches for training.

```python
loader = DataLoader(
    dataset,
    batch_size=32,
    shuffle=True,
    num_workers=0,
)

batch_data, batch_labels = next(iter(loader))

print('Image batch shape:', batch_data.shape)
print('Label batch shape:', batch_labels.shape)
print('Labels:', batch_labels[:10])
```
