```python
import pyg4ometry
```

```python

# 1. Load the GDML file
# This parses the geometry, materials, and solids into a pyg4ometry registry
reader = pyg4ometry.gdml.Reader("geo.gdml")

# 2. Extract the physical world volume (the outermost container)
world_volume = reader.getRegistry().getWorldVolume()

# 3. Initialize the default VTK-based visualizer
visualizer = pyg4ometry.visualisation.VtkViewer()

# 4. Add the world volume to the rendering scene
visualizer.addLogicalVolume(world_volume)

# 5. Launch the interactive 3D viewer window
#visualizer.view()
```
