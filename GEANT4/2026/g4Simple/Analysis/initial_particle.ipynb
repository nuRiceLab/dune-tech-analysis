```python
evtid=GPS["evtid"].array().to_numpy() # Mev to ev
#print(evtid)
```

```python
energy=GPS["kenergy"].array().to_numpy() # Mev to ev 
print(len(energy))
print(f"Max Energy {round(energy.max(),2)} MeV")
print(f"Min Energy {round(energy.min(),2)} MeV")
```

```python
_=plt.hist(energy,bins=40)
plt.xlabel("Energy (MeV)")
```

```python

mx=GPS["mdx"].array().to_numpy() #x 
my=GPS["mdy"].array().to_numpy() #y 
mz=GPS["mdz"].array().to_numpy() #z 
```

```python
plt.title("Momentum Direction")
_=plt.hist2d(mx,my,bins=100)
plt.xlabel("px")
plt.ylabel("py")
```
