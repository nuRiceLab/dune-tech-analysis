```python
import uproot as up
import matplotlib.pyplot as plt
```

```python
f1=up.open("../build/result_t0.root")
```

```python
hits=f1["hits"]
```

```python
hits.keys()
```

```python
print(hits["pname"].array().to_numpy())
```

```python
print(hits["det_name"].array().to_numpy())
```

```python
evtid=hits["evtid"].array().to_numpy()

#print(evtid)
```

```python
energy=hits["edep"].array().to_numpy() # Mev
print(len(energy))
print(f"Max Energy {round(energy.max(),5)} MeV")
print(f"Min Energy {round(energy.min(),6)} MeV")
```

```python
_=plt.hist(energy,bins=40)
plt.xlabel("Energy (MeV)")
```

```python
x=hits["x"].array().to_numpy()
y=hits["y"].array().to_numpy()
_=plt.hist2d(x,y,weights=energy,bins=300)
plt.colorbar(label="edeps")
plt.xlabel("x(cm)")
plt.ylabel("y(cm)")
```

```python
evtmask=evtid==2
plt.title(" Histogram Trajectory ")
x=hits["x"].array().to_numpy()[evtmask]
y=hits["y"].array().to_numpy()[evtmask]
_=plt.hist2d(x,y,weights=energy[evtmask],bins=80)
plt.colorbar(label="edeps")
plt.xlabel("x(cm)")
plt.ylabel("y(cm)")
```

```python
plt.title("Particle Trajectory ")
plt.plot(x,y,marker="o")
plt.xlabel("x(cm)")
plt.ylabel("y(cm)")
```
