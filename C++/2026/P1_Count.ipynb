```c++
#include <iostream>
```

```c++

int value=0;
int maxAttempt=4;

```

```c++
int secret=0;
std::srand(std::time(0));
secret=std::rand()%10;
// for Loop
for (int i=1; i<=maxAttempt;i++)
 {
    std::cout << " What is your value ? " << std::endl;
    std::cin>> value;
    if(value==secret){
        std::cout << "You Win! Number is " << value << std::endl;
        break;
    } else {
         std::cout << " Try again ! " <<std::endl;
    }
    std::cout << " you have " << maxAttempt- i << " many attempts " << " out of " << maxAttempt <<std::endl;
 }
std::cout << " Value was " << secret << std::endl;
```

```c++
// While Loop
int count=1;
int secret=0;
std::srand(std::time(0));
secret=std::rand()%10;
while(count<=maxAttempt){
    std::cout << " What is your value ? " << std::endl;
    std::cin>> value;  
    if(value==secret){
        std::cout << "You Win! Number is " << value << std::endl;
        break;
    } else {
         std::cout << " Try again ! " <<std::endl;
    }
    
    std::cout << " you have " << maxAttempt- count << " many attempts " << " out of " << maxAttempt <<std::endl;
    count++;
}
std::cout << " Value was " << secret << std::endl;
```
