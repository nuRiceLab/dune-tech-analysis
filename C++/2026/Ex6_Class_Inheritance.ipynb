```c++
// Inheritance
#include <iostream>
```

```c++
class person{
  public:
    person(std::string name){fname=name;};
    std::string fname;
    void greet(){std::cout << "Welcome, " << fname << std::endl;}
    void test();
}
```

```c++


void person::test(){
    std::cout << " Test " << std::endl;
}
```

```c++
person p1("james");
//p1.name="james";
p1.greet();
```

```c++
p1.test();
```

```c++
class student :public person {
    public:
        student(std::string name) : person(name) {}
        int student_id=0 ;
        void print_sid(){ std::cout <<"Your student id is " << student_id << std::endl;}
}
```

```c++
student s1("ilker");
s1.student_id=1;
```

```c++
s1.greet();
s1.print_sid();
```

```c++
s1.test();
```
