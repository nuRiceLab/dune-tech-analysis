```c++
#include<iostream>
#include<vector>
// Arrays 
int age[3]={10,20,30};// Accessing the last element

// Accessing the elements
std::cout << age[0] <<std::endl;

// Accessing the last element
std::cout << "Last Element " << std::endl;
std::cout << age[std::size(age)-1] << std::endl;
```

```c++
// Vectors
std::vector<int> vage = {10,20,30};

// Accessing the elements
std::cout << vage[0] <<std::endl;
// Accessing the elements
std::cout << vage.at(0) <<std::endl;

// Accessing the last element
std::cout << "Last Element " << std::endl;
std::cout << vage.at(vage.size()-1) << std::endl;
```

```c++
std::vector<int> v={};

if (v.size()==0){
    v.push_back(10);
    std::cout << "if is executed " << std::endl;
    std::cout << v.at(0) << std::endl;
}
else {
    std::cout << "else is executed " << std::endl;
    std::cout << v.at(0) << std::endl;
}
```
