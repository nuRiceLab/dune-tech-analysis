```c++
#include <iostream>
// Assignment 
```

```c++
// Example for Assigning by a value
std::cout << "Example - Assign by a value " << std::endl;

int a,b;
a=1;
b=a;
std::cout<<" a is " << a << std::endl;
std::cout<<" b is " << b << std::endl;
a=2;
std::cout<<" After changing a" << std::endl;
std::cout<<" a is " << a << std::endl;
std::cout<<" b is " << b << std::endl;
```

```c++
// Example for Assigning by a reference
std::cout << "Example - Assign by a reference " << std::endl;
int a;
a=1;
int &b=a; // References assigned at the declartions
std::cout<<" a is " << a << std::endl;
std::cout<<" b is " << b << std::endl;
a=2;
std::cout<<" After changing a" << std::endl;
std::cout<<" a is " << a << std::endl;
std::cout<<" b is " << b << std::endl;
```

```c++
// Example for Assigning by a reference
std::cout << "Example - Assign by a pointer " << std::endl;
int a;
a=1;
int *b=&a; // References assigned at the declartions
std::cout<<" a is " << a << std::endl;
std::cout<<" b is " << *b << std::endl;
a=2;
std::cout<<" After changing a" << std::endl;
std::cout<<" a is " << a << std::endl;
std::cout<<" b is " << *b << std::endl;
```
