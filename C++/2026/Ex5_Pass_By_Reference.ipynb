```c++
#include <iostream>
```

```c++
// Example - Pass by a value 
void add(int v){
    std::cout << "v is " << v << std::endl;
    v+=20;
    std::cout << "v is " << v << std::endl;
}
```

```c++
std::cout << "Example - Pass by a value" <<std::endl;
int a=1;
std::cout << "a is " << a <<std::endl;
std::cout << "--- After the function ---" << std::endl;
add(a);
std::cout << "a is " << a <<std::endl;
```

```c++
// Example - Pass by a reference 
void addref(int &v){
    std::cout << "v is " << v << std::endl;
    v+=20;
    std::cout << "v is " << v << std::endl;
}
```

```c++
std::cout << "Example - Pass by a reference" <<std::endl; 

int a=1;
std::cout << "a is " << a <<std::endl;
std::cout << "--- After the function ---" << std::endl;
addref(a);
std::cout << "a is " << a <<std::endl;
```

```c++
// Example - Pass by a pointer 
void addpoint(int *v){
    std::cout << "v is " << *v << std::endl;
    *v+=20;
    std::cout << "v is " << *v << std::endl;
}
```

```c++
std::cout << "Example - Pass by a pointer" <<std::endl;
int a=1;
std::cout << "a is " << a <<std::endl;
std::cout << "--- After the function ---" << std::endl;
addpoint(&a);
std::cout << "a is " << a <<std::endl;
```
