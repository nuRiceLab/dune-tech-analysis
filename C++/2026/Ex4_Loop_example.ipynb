```c++
// For Loop Usage
#include <iostream>
using namespace std;
for(int i=0;i<4;i++)
    std::cout << " i is " << i << std::endl;
    
```

```c++
// While Loop
int i = 0 ;
while (i<4){
    std::cout << " i is " << i << std::endl;
    i++;
}
```

```c++
// Do while Loop

int i = 0 ;
do {
    std::cout << " i is " << i << std::endl;
    i++;
} while((i<4))
```
