```c++
class Calculator{
    public:
        double v1, v2;
        std::string op;
        double addition();
        double substraction();
        double division();
        double multiplication();
        void calculate();
    private:
        double result;
};

```

```c++
inline double Calculator::addition(){
    std::cout << "Adding " << v1 << " and " << v2 << std::endl;
    return (v1+ v2);
}

```

```c++
inline double Calculator::multiplication(){
    std::cout << "Multiplying " << v1 << " by " << v2 << std::endl;
    result=(v1*v2);
    return result;
}
```

```c++
inline void Calculator::calculate(){
    double result=0;
    if(op=="+") result=addition();
    else if(op=="*") result=multiplication();
    else {
        std::cout << "Cant Compute " << op << std::endl;
    }
    if(result>0) std::cout << "Your result is " << result <<std::endl;
}
```
